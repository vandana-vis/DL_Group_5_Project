Function definition

```
#Defining a function that takes model and image path as input and prints the predicted image label (Dam/NoDam) along with the Image
def image_lowheaddam_pred(model, image_path):
  img = load_img(image_path) #loading image
  img_arr = img_to_array(img) #converting image to array format
  img_arr = np.reshape(img_arr, (1,500,500,3)) #reshaping the array
  datagen = ImageDataGenerator(rescale=1./255) #Creating an ImageDataGenerator object
  trial_X = datagen.flow(img_arr,batch_size=1) #Loading the image array through datagenerator object
  pred = model.predict(trial_X) #Predicting the label for the image
  plt.imshow(img) #Displaying the image
  predicted_class = ["Dam" if x<=0.5 else "NoDam" for x in pred] #Assigning the label based on probability threshold of 0.5
  print("Predicted class:", predicted_class) #Printing predicted class
```

Loading the model

```
model = tf.keras.models.load_model('/content/drive/My Drive/DLProject/mymod/source_code12/saved_model_face') #loading the saved model
```
